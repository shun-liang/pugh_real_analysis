## Language

### Definition: Single set

A **singleton set** is a set that has exactly one member.

### Notation note: $\subset$

In this book $A \subset B$ means $A$ is a subset of $B$. While in some other literature $A \subset B$ denotes $A$ is a *proper subset* of $B$.

Thus in this book $A \subset A$ is acceptable.

### Definition: Disjoint sets

If $A \cap B = \emptyset$, then $A$ and $B$ are **disjoint**.

### Definition: Symmetric difference

The **symmetric difference** of $A$ and $B$ is the set $A \triangle B$ each of whose elements belongs to $A$ but not to $B$, or belongs to $B$ but not to $A$.

### Definition: Class

A **class** is a collection of sets. The sets are members of the class.

For example, we could consider the class $\varepsilon$ of sets of even natural numbers. Then
- $\{2, 15\}$ is not a member of $\varepsilon$.
- $\{6\}$ is a member of $\varepsilon$.
- $\emptyset$ is a member of $\varepsilon$ as each element of the empty set is even.

If each member of class $A$ is also a member of class $B$, then we call $A$ is a **subclass** of $B$.

For example, the class $\tau$ of sets positive integers divisible by $10$. Then $\tau$ is a subclass of $\varepsilon$, and we write $\tau \subset \varepsilon$.

Another example, let $\mathscr{S}$ be the class of singleton subsets of $\mathbb{N}$, and $\mathbb{D}$ be the class of subsets of $\mathbb{N}$ each of which has exactly two elements. Thus $\{10\} \in \mathscr{S}$, and $\{2, 6\} \in \mathscr{D}$. By definition, $\mathscr{S}$ is not a subclass of $\mathbb{D}$. However, for each member of $\mathscr{S}$, there exist member of $\mathbb{D}$ which the former member is the subset of the latter member.

### Equivalence

#### Definition: Equivalence relation

An **equivalence relation** on a set $S$ is a relation $s \sim s^\prime$ that holdes three properties: For all $s, s^\prime, s^{\prime\prime} \in S$:
- (a) Reflexitivity: $s \sim s$.
- (b) Symmetricity: $s \sim s^\prime \implies s^\prime \sim s$.
- (c) Translativity: $(s \sim s^\prime) \land (s^\prime \sim s^{\prime\prime}) \implies s \sim s^{\prime\prime}$

### Truth

When is a mathematical statement accepted as true?

Things help mathematical statements being accepted as true:
- A picture may be vital in getting you to believe a statement.
- An analogy with something you know to be true may help you understand it.
- ~An authoritative teacher may force you to parrot it.~

A more mathematical answer:

> Only when it has a proof inside a familiar mathematical framework.

A formal proof is the ultimate and only reason to accept a mathematical statement as true.
